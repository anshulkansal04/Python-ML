```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('filtered.csv')

# Dictionary to map courses to their abbreviations
course_abbreviations = {
    "B.Tech Mechanical Engineering": "BT_ME",
    "B.Tech Computer Science and Engineering": "BT_CSE",
    "B.Tech Electrical Engineering": "BT_EE",
    "B.Tech Civil Engineering": "BT_CE",
    "B.Tech Electronics and Communication Engineering": "BT_ECE",
    "B.Tech Chemical Engineering": "BT_ChemE",
    "M.Tech Chemical Engineering": "MT_ChemE",
    "M.Tech Communication and Networks": "MT_CN",
    "M.Tech Computer Science": "MT_CS"
}

# Function to filter and replace courses with their abbreviations
def filter_and_replace_courses(courses):
    if isinstance(courses, str):  # Only process if the value is a string
        filtered_courses = [
            course_abbreviations[course.strip()]
            for course in courses.split(',')
            if course.strip() in course_abbreviations
        ]
        return ', '.join(filtered_courses) if len(filtered_courses) >= 6 else ''
    return ''  # Return an empty string if it's not a valid course entry

# Apply the function to the 'courses' column
df['Courses'] = df['Courses'].apply(filter_and_replace_courses)

# Remove rows where 'courses' column is empty (less than 6 courses)
df = df[df['Courses'] != '']

# Save the filtered dataframe with abbreviations to a new CSV file
df.to_csv('filtered_courses_abbr.csv', index=False)

# Print the filtered dataframe
print(df)
```
